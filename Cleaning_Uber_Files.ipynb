```python
import pandas as pd
import numpy as np
```

```python
uber = pd.read_csv('UberFaresCleaned.csv')
```

```python
uber = uber[['Zip_Codes', 'Destination_Airport',
       'Uber_X_Evening', 'Uber_XL_Evening', 'Uber_Ski_Evening',
       'Uber_X_Morning', 'Uber_XL_Morning', 'Uber_Ski_Morning',
       'Uber_X_Afternoon', 'Uber_XL_Afternoon', 'Uber_Ski_Afternoon',
       'Travel_Times_(Min)', 'Distance_(Mi)']]
```

```python
uber_x = uber[['Zip_Codes','Destination_Airport',
      'Travel_Times_(Min)','Distance_(Mi)','Uber_X_Morning',
     'Uber_X_Afternoon','Uber_X_Evening']]

```

```python
uber_x_melted = pd.melt(uber_x,id_vars=['Zip_Codes','Destination_Airport',
                        'Travel_Times_(Min)','Distance_(Mi)'],value_vars=['Uber_X_Morning',
     'Uber_X_Afternoon','Uber_X_Evening'],var_name='Time_Day',value_name='Fare')
```

```python
uber_x_melted
```

# Now do for uber xl

```python
uber_xl = uber[['Zip_Codes','Destination_Airport',
      'Travel_Times_(Min)','Distance_(Mi)','Uber_XL_Morning',
     'Uber_XL_Afternoon','Uber_XL_Evening']]
uber_xl_melted = pd.melt(uber_xl,id_vars=['Zip_Codes','Destination_Airport',
                        'Travel_Times_(Min)','Distance_(Mi)'],value_vars=['Uber_XL_Morning',
     'Uber_XL_Afternoon','Uber_XL_Evening'],var_name='Time_Day',value_name='Fare')
```

```python
uber_xl_melted.head()
```

## Now do for uber ski

```python
uber_ski = uber[['Zip_Codes','Destination_Airport',
      'Travel_Times_(Min)','Distance_(Mi)','Uber_Ski_Morning',
     'Uber_Ski_Afternoon','Uber_Ski_Evening']]
uber_ski_melted = pd.melt(uber_ski,id_vars=['Zip_Codes','Destination_Airport',
                        'Travel_Times_(Min)','Distance_(Mi)'],value_vars=['Uber_Ski_Morning',
     'Uber_Ski_Afternoon','Uber_Ski_Evening'],var_name='Time_Day',value_name='Fare')
```

```python
uber_ski_melted.head()
```

## Merge on zip codes and destination ariport key

```python
#first merge x and xl
a = uber_x_melted.merge(uber_xl_melted,how='outer',on = ['Zip_Codes','Destination_Airport',
                                                    'Travel_Times_(Min)','Distance_(Mi)','Fare',
                                                    'Time_Day'])
#merge with ski
final = a.merge(uber_ski_melted,how='outer',on = ['Zip_Codes','Destination_Airport',
                                                    'Travel_Times_(Min)','Distance_(Mi)','Fare',
                                                    'Time_Day'])
```

```python
#def split to get time of day
def split_time_day(x):
    split_elements = x.split("_")
    return(split_elements[2])

def split_type(x):
    split_elements = x.split("_")
    return(split_elements[1])
```

```python
final['Time'] = final['Time_Day'].apply(split_time_day)
```

```python
final['Uber_Type'] = final['Time_Day'].apply(split_type)
```

```python
final.to_csv('Uber Melted.csv')
```
